1. LLaMA3 ê¸°ë°˜ í…ìŠ¤íŠ¸ ìƒì„± í•¨ìˆ˜

````python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "meta-llama/Meta-Llama-3-8B-Instruct"

# ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (ìµœì´ˆ ì‹¤í–‰ ì‹œ ì‹œê°„ ì†Œìš”)
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=True)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")


def extract_email_body(full_text, prompt):
    # í”„ë¡¬í”„íŠ¸ ê¸¸ì´ë§Œí¼ ìë¥´ê³ , ì•ë’¤ ê³µë°± ì œê±°
    return full_text[len(prompt):].strip()

def generate_email(name, context):
    # í”„ë¡¬í”„íŠ¸ ì •ì˜
    prompt = f"""
    ì•„ë˜ ì¡°ê±´ì— ë§ì¶° ê³ ê°ì—ê²Œ ë³´ë‚¼ ê°ì‚¬ ì´ë©”ì¼ì„ í•œê¸€ë¡œ ì‘ì„±í•´ì¤˜:

    ê³ ê° ì´ë¦„: {name}
    ìƒí™©: {context}
    ì´ë©”ì¼ ë‚´ìš©:
    """
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=200, temperature=0.7)
    full_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # ë©”ì‹œì§€ ë¶€ë¶„ë§Œ ì¶”ì¶œ
    return extract_email_body(full_text, prompt)
````

 2. ê³ ê° ëª©ë¡ ë° ì´ë©”ì¼ ë‚´ìš© ìƒì„±

````python
import pandas as pd

# ê³ ê° ì •ë³´
df = pd.DataFrame({
    "ì´ë¦„": ["ì •ì¸"],
    "ì´ë©”ì¼": ["jikim@imguru.co.kr"],
    "ìƒí™©": ["ì„œë¹„ìŠ¤ ê°€ì… ì™„ë£Œ"]
})

# ê° ê³ ê°ì—ê²Œ ì´ë©”ì¼ ë‚´ìš© ìƒì„±
df["ë³¸ë¬¸"] = df.apply(lambda row: generate_email(row["ì´ë¦„"], row["ìƒí™©"]), axis=1)
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

def generate_llama3_summary(input_text):
    model_name = "meta-llama/Meta-Llama-3-8B-Instruct"

    # ëª¨ë¸ & í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸°
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,
        device_map="auto"
    )

    # í”„ë¡¬í”„íŠ¸ êµ¬ì„±
    prompt = f"""
    ë‹¤ìŒ ë‚´ìš©ì„ ìš”ì•½í•˜ì—¬ ë¹„ì¦ˆë‹ˆìŠ¤ ë¶„ì„ ë³´ê³ ì„œ ì´ˆì•ˆì„ ì‘ì„±í•´ì¤˜.

    "{input_text}"

    ë³´ê³ ì„œ:
    """

    # í† í°í™” ë° ìƒì„±
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        temperature=0.7,
        do_sample=True,
        top_p=0.9
    )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # ë³´ê³ ì„œ ë¶€ë¶„ë§Œ ì¶”ì¶œ
    if "ë³´ê³ ì„œ:" in result:
        return result.split("ë³´ê³ ì„œ:")[-1].strip()
    return result.strip()
````
